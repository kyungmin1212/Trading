```python
import datetime
import pandas as pd
import numpy as np
import plotly.graph_objects as go

import warnings
warnings.filterwarnings(action='ignore')

file_name = './BTCUSDT_15m_2020-01-01-00-00-00_2022-12-31-23_59_00.csv'
df = pd.read_csv(file_name)
df
```

```python
from ta.trend import SMAIndicator


df['sma50'] = SMAIndicator(df['close'],window=50).sma_indicator()
df['sma200'] = SMAIndicator(df['close'],window=200).sma_indicator()

df = df.dropna()
df
```

```python
df['sma50>sma200']=df['sma50']>df['sma200']
df
```

```python
df['check']=df['sma50>sma200'].shift(1).fillna(0)
df
```

```python
def get_up_down(df):
    if df['sma50>sma200']==True and df['check']==False:
        ans = 'Up'
    elif df['sma50>sma200']==False and df['check']==True:
        ans = 'Down'
    else:
        ans = 'same'
    return ans

df['cross_check']= df.apply(lambda x:get_up_down(x),axis=1)
df
```

```python
# 구매 수량 계산(BTC 기준, 다른 코인이라면 https://www.binance.com/en/trade-rule 를 참고하여 100000 숫자 조정 필요)
def cal_amount(usdt_balance,cur_price,portion,leverage):
    """
    거래시점 기준 현재 자산 대비 portion 만큼의 금액을 cur_price에 구매할때 구매량 구하기
    :params float usdt_balance : 현재 시점 usdt_balance
    :params float cur_price : 현재 시점 BTC(또는 다른 코인) 가격
    :params float portion : 자산대비 구매비율
    :params int leverage : 레버리지(1이상, 1이면 그냥 자기 자산만으로 거래)
    :return amount ex) BTC/USDT라면 0.00346 (단위 BTC)
    :rtype float
    """
    usdt_trade = usdt_balance*leverage * portion
    amount = math.floor((usdt_trade*100000)/cur_price)/100000 # 최소 단위가 0.00001 BTC이므로
    return amount

# btc를 통해 usdt계산하기
def cal_btc_to_usdt(cur_price,btc_amount):
    return cur_price*btc_amount

# 체결된 btc수량을 통해 수수료 구하기
def cal_trade_fee(trade_type,cur_price,btc_amount):
    """
    거래시점 기준 체결된 BTC 수량의 수수료 구하기
    :params str trade_type : 'market' or 'limit' / 시장가 or 지정가
    :params float cur_price : 현재 시점 BTC(또는 다른 코인) 가격
    :params float btc_amount : btc 보유량
    :return cur_trade_fee ex) BTC/USDT라면 0.00915 (단위 usdt)
    :rtype float
    """
    if trade_type == 'market':
        fee = 0.0004 # 0.04 %
    elif trade_type == 'limit':
        fee = 0.0002 # 0.02 %
    
    usdt = cur_price*btc_amount
    cur_trade_fee = usdt*fee
    
    return cur_trade_fee
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df_split = df[:1000]

fig=make_subplots(rows=2, cols=1, shared_xaxes=True, row_heights=[0.3,0.7])

candle = go.Candlestick(name ='ohlc',
                        x=df_split['datetime'],
                        open=df_split['open'],
                        high=df_split['high'],
                        low=df_split['low'],
                        close=df_split['close'])

sma50= go.Scatter(name ='sma50',x=df_split['datetime'], y=df_split['sma50'])
sma200 = go.Scatter(name ='sma200',x=df_split['datetime'], y=df_split['sma200'])


volume = go.Bar(name='volume',
                x=df_split['datetime'],
                y=df_split['volume'])

fig.add_trace(volume,row=1,col=1)
fig.add_trace(candle,row=2,col=1)

fig.add_trace(sma50,row=2,col=1)
fig.add_trace(sma200,row=2,col=1)


# up , down 표시
for idx in df_split.index:
    if df_split.loc[idx,'open_close_check']==1:
        fig.add_annotation(
                x=df_split.loc[idx,'datetime'],
                y=df_split.loc[idx,'close'],
                text='C',row=2,col=1
                )

fig.show()
```
